### Part A - Representations of functions

```python
x = np.arange(-1, 0.9, 0.01)
f = 5 + 2*x + x**2   # coefficients 5, 2 and 1

plt.figure(figsize=(5,4))
plt.plot(x,f)
```

```python
sx = np.zeros(round(len(x)/20.0))
sf = np.zeros(round(len(x)/20.0))
print(sx)
print(sf)
```

```python
j = 0
for i in range(0, len(x)+1, 21):
    print("%.2f, %.2f" % (x[i], f[i]))
    sx[j] = x[i]
    sf[j] = f[i]    
    j = j+1
```

```python
Y = sf
N = 2 # degree of polynomial

A = np.zeros( [ len(Y), N+1] )
A[:,0] = sx**N
A[:,1] = sx**(N-1)
A[:,2] = 1
print(A)
```

```python
from numpy.linalg import inv

C1 = inv(np.matmul(np.transpose(A), A))
C2 = np.matmul(np.transpose(A), Y)
C = np.matmul(C1, C2)
print(C)
```

### Part B - Sinusoids and Complex Exponentials

Combining sine and cosine into a complex exponential value

Remember we want functions with the property:

$f(t) = f(t+ k T)$,
in which $T$ is the period and $k=0,1,\ldots$

```python
t = 2.5
print("%.3f" % (np.sin(t)))

T = 2*np.pi
print("%.3f" % (np.sin(t+T)))
```

```python
for i in range(5):
    print("%d %.3f" % (i, np.sin(t+T*i)))
```

```python
for i in range(5):
    print("%d %.3f" % (i, np.cos(t+T*i)))
```

```python
t = np.arange(0, T*5, 0.1)

fcos = np.cos(t)
fsin = np.sin(t)

plt.figure(figsize=(9,3))
plt.plot(t, fcos)
plt.plot(t, fsin)
```

```python
plt.figure(figsize=(4,4))
plt.plot(fcos, fsin)
```

```python
fcos = 2*np.cos(t)
fsin = 4*np.sin(t)
plt.plot(fcos, fsin)
```

```python
plt.figure(figsize=(9,3))
plt.plot(t, fcos)
plt.plot(t, fsin)
```

To create a complex number in Python we use the letter "j"

```python
complex_number = (3 + 2j)
print(complex_number)
```

```python
print(complex_number.real)
print(complex_number.imag)
```

```python
expon_2_5 = np.exp(2.5j)
print(expon_2_5)
```

```python
print("Real part: %.4f, Imaginary part = %.4f" % (expon_2_5.real, expon_2_5.imag))
```
